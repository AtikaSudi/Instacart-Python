### Step 3

```python
# Histogram for order_hour_of_day
```

```python
#Histogram for order_hour_of_day column with 24 bins because there are only 24 hours in a day
hist_3 = df_combined['time_ordered'].plot.hist(bins =24)
```

```python
hist_3.figure.savefig('..\Analysis\Visualisations\hist_3_time_orderd.png')
```

The histogram illustrates the distribution of shoppers throughout the day, showing the lowest number of shoppers between midnight and 6 AM, reaching its highest point between 10 AM and 4 PM, and gradually declining from 4 PM until 11 PM. The majority of orders start at the seventh hour, reach their peak at the tenth hour, and decrease after the fifteenth hour.
