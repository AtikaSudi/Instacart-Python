### If-Statements with User-Defined Functions

```python
# Define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# max price check
df['prices'].max()
```

### If-Statements with the loc() Function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

### If-Statements with For-Loops

```python
df_merged['order_dow'].value_counts(dropna = False)
```

```python
result = []

for value in df_merged["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
df_merged['busiest_day']= result
```

```python
df_merged['busiest_day'].value_counts(dropna = False)
```

```python
time_ordered = []

for value in df_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    time_ordered.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    time_ordered.append("Fewest orders")
  else:
    time_ordered.append("Average orders")
```
